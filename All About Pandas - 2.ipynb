> <b>Outlier Treatment</b><br>
> * An outlier is a data point that is far away from all the other data points 


#### IQR METHOD FOR OUTLIER DETECTION
Steps
1. Arrange the data in increasing order<br>
2. Calculate Q1 and Q3<br>
3. Find interquartile range (Q3-Q1)<br>
4. Find lower bound Q1*1.5<br>
5. Find upper bound Q3*1.5<br>
> Anything that lies outside of lower and upper bound is an outlier

```python
data = pd.read_csv("weight-height.csv")
data.head(5)
```

```python
data.describe().T
# as we can see that the max value is a lot higher than 75% 
```

```python
# Task - 1 Find outlier from height .
q1_height = data.Height.quantile(0.25)
q3_height = data.Height.quantile(0.75)
iqr_height = q3_height-q1_height
iqr_height
```

```python
lc_height = q1_height - 1.5*iqr_height
uc_height = q3_height + 1.5*iqr_height
```

```python
uc_height,lc_height
```

```python
data[(data.Height<lc_height)|(data.Height>uc_height)]
# these are all the outlier values 
```

```python
# Task - 2 remove the outliers
data.Height.clip(lower = lc_height,upper=uc_height)
```

#### Binning / Grouping 
* Converting Continuous/Numeric data into Categorical/Character data.Example : [Age] - Teenager,Adult,Senior,Senior-citizen
* We can use method <<<b> cut(),qcut() and np.where </b>>> for this task
* When class intrval is same its called bin and when it is done conditionally they are called groups

```python
# We can use np.where to make the groups as per our conditions and it works just like if -else staement
Orders["Sales_Category"] = np.where((Orders.Sales<5000),"Low",
                            np.where((Orders.Sales>=5000)& (Orders.Sales< 10000),"Medium",
                             np.where((Orders.Sales>=10000)& (Orders.Sales< 15000),"High",
                              np.where((Orders.Sales>=15000),"Very-High",""))))
```

```python
Orders[["State","Sales","Sales_Category"]]
```

```python
# pd.cut is used to create bins on the series
pd.cut(Orders.Sales,10)
# here the data is divided into 10 equal parts from minimum to maximum
```

```python
# this is just same like np.where 
pd.cut(Orders.Sales,range(0,26000,5000))
```

```python
# we can create groups in the data by providing the class-interval 
pd.cut(Orders.Sales,[0,2000,4000,7000,18000,50000])
```

```python
# pd.qcut will split the data based on quantiles
pd.qcut(Orders.Sales,10) # cutting on deciles
```

```python
pd.qcut(Orders.Sales,4) # distributing data into 4 quartiles
```

#### Grouping of the data / Summaries
* We use group by to split the data into groups 
* Applying a function on the data
* Combining groups into a dataframe

```python
a = Orders.groupby("Country")
```

```python
a.sum()
```

```python
# find the totalsales of each Market
Orders.groupby("Market").Sales.sum().round(2)
# here we have used aggregation on the data
```

```python
# find the total and average sales in each Country
Orders.groupby("Country").Sales.agg(['mean','sum'])
```

```python
# find the total sales and average sales of each state and each segment
Orders.groupby(["State","Segment"]).Sales.agg(["sum","mean"]).reset_index()
```

```python
# find the total and average of sales and profit of each state
Orders.groupby("State")[["Sales","Profit"]].agg({"Sales":[np.sum,np.mean],"Profit":[np.sum,np.mean]}).reset_index()
# we ca pass arguments in the form of dictionary to form diffrent operations
```

```python
# get the total of the sales and average of the profit of each state
Orders.groupby("State")[["Sales","Profit"]].agg({"Sales":[np.sum],"Profit":[np.mean]}).reset_index()
```

```python
# give a df with top 10 states based on the sales
a = pd.DataFrame(Orders.groupby("State").Sales.sum().sort_values(ascending=False).head(10).reset_index())
a
```

### Data Combination
* Horizontally [Row-wise] or Vertically[column-wise]

#### 1. Joins/Merge
* Adding one or more than one column from one table to another table based on common variable
* To join we need to understand the relationship in the data in both tables.
<ol>
<li>One to One</li>
<li>One to Many</li>
<li>Many to one</li>
<li>Many to Many</li>
</ol>


```python
df1 = pd.DataFrame({"Student_Name":["Seb","Rick","Morty","Thomas"],"Student_id":[1001,1002,1003,1004]})
df2 = pd.DataFrame({"Student_Name":["nick","Rick","Morty"],"Subjects":["Animation","Music","Sports"]})
```

```python
# we can use pd.merge to join two tables
pd.merge(left=df1,right=df2,left_on="Student_Name",right_on="Student_Name",how = "inner")
```

```python
pd.merge(left=df1,right=df2,left_on="Student_Name",right_on="Student_Name",how = "left")
```

```python
pd.merge(left=df1,right=df2,left_on="Student_Name",right_on="Student_Name",how = "right")
```

```python
pd.merge(left=df1,right=df2,left_on="Student_Name",right_on="Student_Name",how = "outer")
```

#### 2. Appending
* Adding the rows(table) right below another table (Union in sql)

```python
a = Orders.sample(n = 5,random_state=123)
b = Orders.sample(n = 15,random_state=1234)
c = Orders.sample(n = 4,random_state=1235)
d = Orders.sample(n = 8,random_state=1236)
```

```python
# as they are coming from same entity so we can append them 
Final_df = a.append([b,c,d], ignore_index = True)
Final_df.head(5)
```

```python
# same thing we can do with concat and axis = 0 is for horizontal appending
e = pd.concat([a,b,c,d],ignore_index=True,axis = 0)
e
```
